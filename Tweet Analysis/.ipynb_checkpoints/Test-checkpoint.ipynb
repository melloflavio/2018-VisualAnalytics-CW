```python
import geopandas as gpd
from shapely.geometry import Polygon
```

```python
from lxml import objectify

main = objectify.parse('London_division_2500-150.xml')
```

```python
lxRoot = main.getroot()
all_polygons = []
all_poly_lists = []
i = 0
for obj in lxRoot.iterchildren():
    i = i+1
    shape_points = []
    for point in obj.shape.iterchildren():
        shape_points.append((point.xCoord, point.yCoord))
    polygon_geom = Polygon(shape_points)
#     print(polygon_geom)
    crs = {'init': 'epsg:4326'}
    polygon = gpd.GeoDataFrame(index=[i], crs=crs, geometry=[polygon_geom])       
#     print(polygon.geometry)
#     polygon = gpd.GeoDataFrame(index=[0], geometry=[polygon_geom])       
    all_polygons.append(polygon)
```

```python
from lxml import objectify

main = objectify.parse('London_division_2500-150.xml')
lxRoot = main.getroot()
allPlaces = []
for obj in lxRoot.iterchildren():
    place = {}
    place['id'] = int(obj.id)
    place['name'] = str(obj.name)
    
    shape_points = []
    for point in obj.shape.iterchildren():
        shape_points.append((point.xCoord, point.yCoord))
    polygon_geom = Polygon(shape_points)
    place['polygon'] = polygon_geom
    
    place['value'] = int(obj.id)
#     print(place)
    allPlaces.append(place)

myDf = pd.DataFrame.from_records(allPlaces)
myDf.head()
```

```python
gdf = geopandas.GeoDataFrame(myDf, geometry='polygon')
gdf.head()
```

```python

gdf.to_file('FullPlaces.json', driver="GeoJSON")
```
